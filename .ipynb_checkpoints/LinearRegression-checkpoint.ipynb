```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

import statsmodels.api as sm

%matplotlib inline
```

```python
data = pd.read_csv('data/Advertising.csv', index_col=0)
data.head()
```

## Simple Linear Regression

```python
plt.figure(figsize=(16,8))
plt.scatter(data['TV'], data['sales'], c='black')
plt.xlabel('Dinero gastado en avisos de TV (US$)')
plt.ylabel('Ventas (US$)')
plt.show()
```

```python
X = data['TV'].values.reshape(-1,1)
y = data['sales'].values.reshape(-1,1)

reg = LinearRegression()
reg.fit(X,y)

print(f'El modelo linear es: \n y = {reg.intercept_[0]} + {reg.coef_[0][0]} * TV')
```

```python
predictions = reg.predict(X)

plt.figure(figsize=(16,8))
plt.scatter(X, y, c='black')
plt.plot(X, predictions, c='blue', linewidth=2)
plt.xlabel('Dinero gastado en avisos de TV (US$)')
plt.ylabel('Ventas (US$)')
plt.show()
```

```python
X = data['TV']
y = data['sales']

exog = sm.add_constant(X)
est = sm.OLS(y, exog).fit()

print(est.summary())
```

## Multiple LInear Regression

```python
Xs = data.drop(['sales'], axis=1)
y = data['sales'].values.reshape(-1,1)

reg = LinearRegression()
reg.fit(Xs, y)

print(f'El modelo linear es: \n Y = {reg.intercept_[0]} + {reg.coef_[0][0]}*TV + {reg.coef_[0][1]}*radio + {reg.coef_[0][2]}*newspaper')
```

```python
X = np.column_stack((data['TV'], data['radio'], data['newspaper']))
y = data['sales'].values.reshape(-1,1)

exog = sm.add_constant(X)
est = sm.OLS(y, exog).fit()

print(est.summary())
```
